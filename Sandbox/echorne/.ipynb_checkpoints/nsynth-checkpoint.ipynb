```python
from os import walk
```

```python
acousticGuitars = []
musicFolder = '../../'
testFolder = 'nsyth_test/audio/'
for root, dirs, files in walk(musicFolder+testFolder):
    for file in files:
        if file.startswith('guitar_acoustic') and file.endswith('.wav'):
            acousticGuitars.append(file)
    break
acousticGuitars[:20]
```

```python
import scipy
```

The following data.shape is 1d because the audio from the source is mono.

```python
from scipy.io import wavfile
import scipy.io

samplerate, data = wavfile.read(musicFolder+testFolder+acousticGuitars[0])
print(f"sample size and channels = {data.shape}")

length = data.shape[0] / samplerate
print(f"length = {length}s")
```

```python
import matplotlib.pyplot as plt

import numpy as np

time = np.linspace(0., length, data.shape[0])

plt.plot(time, data[:], label="mono")

plt.legend()

plt.xlabel("Time [s]")

plt.ylabel("Amplitude")

plt.show()
```

If we get data with stereo sound it could affect the way it's interpreted. It would probably help to specifically analyze each channel separately.

```python
samplerate, data = wavfile.read(musicFolder+'181425__serylis__guitar-chord.wav')
print(f"sample size and channels = {data.shape}")

length = data.shape[0] / samplerate
print(f"length = {length}s")
```

```python
time = np.linspace(0., length, data.shape[0])

plt.plot(time, data[:, 0], label="Left Channel")
plt.plot(time, data[:, 1], label="Right Channel")

plt.legend()

plt.xlabel("Time [s]")

plt.ylabel("Amplitude")

plt.show()
```

```python
import pandas as pd
from os import path
```

```python
examplesJFile = '../../nsyth_test/examples.json'
examples = pd.read_json(examplesJFile).transpose()
examples
```

```python
find_q = pd.DataFrame([[0]*10]*examples.shape[0])
find_q[1] = 1
subset = pd.DataFrame(examples['qualities'].tolist())
comparison_df = find_q[1].eq(subset[1])
subset[comparison_df]
```

```python
qualityColNames = []
for i in range(10):
    find_q = pd.DataFrame([[0]*10]*examples.shape[0])
    find_q[i] = 1
    subset = examples[examples['qualities'] == find_q]
    qualityColNames.append(subset['qualities_str'].iloc[0])
print(qualityColNames)
```
