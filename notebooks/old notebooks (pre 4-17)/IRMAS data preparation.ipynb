# Loop through instruments and store labels and MFCCs in a json file

```python
instruments = ["cel", "cla", "flu", "gac", "gel", "org", "pai",
               "sax", "tru", "vio", "voi"]
drums = ["nod", "dru"]
genre = ["cou_folk", "cla", "pop-rock", "lat-sou"]

# set a boolean variable to break after 1 iteration for debugging
test = False

data = {}

### Hyperparams
n_fft = 2048
hop_length = 512
n_mfcc = 13
notes = """Optional notes not passed."""

# store json dataset description
data['metadata'] = {
    'n_fft': n_fft,
    'hop_length': hop_length,
    'n_mfcc': n_mfcc,
    'notes': notes,
}

# unique key
id = 0

# loop through all filenames for one instrument
for i, filename in enumerate(os.listdir(IRMAS_PATH + instruments[0])):

  # output every 100 files for logging
  if i%100 == 0:
    print("loading file: ", i)
#/content/drive/MyDrive/ITCS 5156 project/IRMAS dataset/IRMAS-TrainingData/cel/[cel][cla]0123__1.wav
  #filename = "[cel][cla]0123__1.wav"
  # get file path
  file_path = IRMAS_PATH + instruments[0] + "/" + filename

  # pull data from file
  signal, sr = librosa.load(file_path, sr=44100)

  # generate MFCC
  mfccs = librosa.feature.mfcc(signal, n_fft=n_fft, hop_length=hop_length,
                               n_mfcc=n_mfcc)

  # get label (get this directly from the filename for the sake of redundancy)

  drum_label = ""
  genre_label = ""

  second_label = filename.split("[")[2].split("]")[0]
  if second_label in drums:
    drum_label = second_label

    # get genre label
    genre_label = filename.split("[")[3].split("]")[0]
  else:
    genre_label = second_label

  print("primary_instrument: ", primary_instrument)
  print("drum_label: ", drum_label)
  print("genre_label: ", genre_label)

  # store data and label
  data[str(id)] = {
      'mfcc': mfccs.tolist(),
      'primary_instrument': primary_instrument,
      'drums': drum_label,
      'genre': genre
  }
  id += 1
  # break out of the loop if we're debugging
  if test:
    # plot mfcc for testing
    print(mfccs.shape)
    librosa.display.specshow(mfccs, sr=sr, hop_length=hop_length)
    plt.xlabel("time")
    plt.ylabel("mfcc")
    plt.colorbar()
    print("debugging - breaking after 1 iteration")
    break  
```
